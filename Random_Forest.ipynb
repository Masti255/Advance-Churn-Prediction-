```python
def process_and_merge_chunks(data_list, chunk_size):
    """
    Processes a large list of data in chunks, converts each chunk to a DataFrame, and merges them into a single DataFrame.

    Parameters:
    - data_list: The complete list of data.
    - chunk_size: The number of items per chunk to process at a time.

    Returns:
    - A single DataFrame containing all the merged data.
    """
    # Initialize an empty list to store DataFrames
    dfs = []
    
    # Process the data list in chunks
    for i in range(0, len(data_list), chunk_size):
        # Extract a chunk of data
        chunk = data_list[i:i + chunk_size]
        
        # Convert the chunk to a DataFrame
        df_chunk = pd.DataFrame(chunk)
        
        # Append the DataFrame to the list
        dfs.append(df_chunk)
    
    # Concatenate all DataFrames into a single DataFrame
    merged_df = pd.concat(dfs, ignore_index=True)
    
    return merged_df
```

```python
import numpy as np
import pickle
from tqdm import tqdm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Step 1: Split the data into 90% train/validation and 10% test set
X = X_final_top_features  # Features
y = y  # Target variable

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Step 2: Initialize StratifiedKFold for 5-fold cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Step 3: Initialize lists to store metrics for each fold
test_set_accuracies = []
test_set_precisions = []
test_set_recalls = []
test_set_f1_scores = []

# Step 4: Perform the 5-fold cross-validation on the training/validation set
for fold, (train_index, val_index) in enumerate(kf.split(X_train_val, y_train_val)):
    print(f"Fold {fold + 1}")
    
    # Split the training/validation data into training and validation sets
    X_train, X_val = X_train_val.iloc[train_index], X_train_val.iloc[val_index]
    y_train, y_val = y_train_val.iloc[train_index], y_train_val.iloc[val_index]
    
    # Step 5: Initialize the RandomForest model
    model = RandomForestClassifier(
        n_estimators=200,  # Adjust this to your hyperparameter settings
        max_depth=20,
        min_samples_split=10,
        min_samples_leaf=4,
        bootstrap=True,
        random_state=42,
        warm_start=True)  # Allows incremental training of the model
    
    # Incremental training
    for i in tqdm(range(1, model.n_estimators + 1), desc='Training Progress'):
        model.set_params(n_estimators=i)
        model.fit(X_train, y_train)
    
    # Step 6: Validate the model on the validation set
    y_val_pred = model.predict(X_val)
    val_accuracy = accuracy_score(y_val, y_val_pred)
    print(f"Validation Accuracy for Fold {fold + 1}: {val_accuracy:.4f}")
    
    # Step 7: Evaluate the model on the test set
    y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_set_accuracies.append(test_accuracy)

    # Calculate precision, recall, and F1 score for the test set
    labels = np.unique(np.concatenate((y_test, y_test_pred)))
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_test_pred, average='binary', pos_label='1')    
    test_set_precisions.append(precision)
    test_set_recalls.append(recall)
    test_set_f1_scores.append(f1_score)

    print(f"Test Set Accuracy for Fold {fold + 1}: {test_accuracy:.4f}")
    print(f"Test Set Precision for Fold {fold + 1}: {precision:.4f}")
    print(f"Test Set Recall for Fold {fold + 1}: {recall:.4f}")
    print(f"Test Set F1 Score for Fold {fold + 1}: {f1_score:.4f}")
    
    # Step 8: Save the model after each fold
    filename = f'rf_model_fold_{fold + 1}.pkl'
    with open(filename, 'wb') as file:
        pickle.dump(model, file)
    print(f"Model saved as {filename}")

# Step 9: Calculate and print the average test set accuracy, precision, recall, and F1 score across all folds
average_test_accuracy = np.mean(test_set_accuracies)
average_test_precision = np.mean(test_set_precisions)
average_test_recall = np.mean(test_set_recalls)
average_test_f1_score = np.mean(test_set_f1_scores)

print(f"\nAverage Test Set Accuracy across all folds: {average_test_accuracy:.4f}")
print(f"Average Test Set Precision across all folds: {average_test_precision:.4f}")
print(f"Average Test Set Recall across all folds: {average_test_recall:.4f}")
print(f"Average Test Set F1 Score across all folds: {average_test_f1_score:.4f}")
```

```python
import joblib
import numpy as np
from scipy.stats import mode
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Step 1: Load all fold-wise models
models = []
num_folds = 5  # Adjust this to the number of folds you used

for fold in range(1, num_folds + 1):
    model = joblib.load(f'rf_model_fold_{fold}.pkl')
    models.append(model)

# Step 2: Make predictions with each model on the test set
all_predictions = []

for model in models:
    y_test_pred = model.predict(X_test)
    all_predictions.append(y_test_pred)

# Step 3: Convert list of predictions to a numpy array, ensuring numeric type
all_predictions = np.array(all_predictions, dtype=np.int64)

# Step 4: Convert y_test to integer type to match the predictions
y_test_int = y_test.astype(np.int64)

# Step 5: Perform majority voting (find the most common prediction for each instance)
y_test_pred_ensemble, _ = mode(all_predictions, axis=0)
y_test_pred_ensemble = y_test_pred_ensemble.ravel()  # Flatten the result

# Step 6: Evaluate the ensemble's performance
ensemble_accuracy = accuracy_score(y_test_int, y_test_pred_ensemble)
print(f"Ensemble Model Test Accuracy: {ensemble_accuracy:.4f}")

# Optionally calculate precision, recall, and F1 score
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test_int, y_test_pred_ensemble, average='binary')
print(f"Ensemble Model Precision: {precision:.4f}")
print(f"Ensemble Model Recall: {recall:.4f}")
print(f"Ensemble Model F1 Score: {f1_score:.4f}")
```
